## Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

## Read Image

```python
image = cv2.imread('1629614829845.jpg')
```

## Read the Image in RGB format and View

```python
im_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,6))
plt.imshow(im_rgb)
plt.title("Original Image")
plt.axis('off');

```

## Converting the Original Image to Grayscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
plt.figure(figsize=(10,6))
plt.imshow(gray_image, cmap='gray')
plt.title("GrayScale Image")
plt.axis('off');
```

## Inverting Image

```python
inverted_image = 255 - gray_image
plt.figure(figsize=(10,6))
plt.imshow(inverted_image, cmap ='gray')
plt.title('Inverted Image')
plt.axis('off');
```

## Blur the Image

```python
blur = cv2.GaussianBlur(inverted_image,(21,21),0)
plt.figure(figsize=(10,6))
plt.imshow(blur,cmap='gray')
plt.title("Smoothen Image")
plt.axis('off');
```

## Converting Image to Pencil Sketch

```python
sketch = cv2.divide(gray_image, 255 - blur,scale=255)
plt.figure(figsize=(10,6))
plt.imshow(sketch,cmap='gray')
plt.title('Final Sketch Image')
plt.axis('off');
```
